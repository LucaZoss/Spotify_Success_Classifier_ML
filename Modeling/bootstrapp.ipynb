```python
import pandas as pd

data = pd.read_csv("/Users/lucazosso/Desktop/IE_Course/Term_2/Machine Learning II/Group Assignement/data/testing/nano_bootsrapp.csv", index_col="song_id")
data.head()
```

```python
from sklearn.utils import resample

# Aim for a target size that's more representative of a balanced dataset
# This could be based on a median class size or a specific percentile.
target_size = data['track_popularity'].value_counts().median()

# Initialize an empty DataFrame to hold the resampled data
resampled_data = pd.DataFrame()

for popularity_score, group in data.groupby('track_popularity'):
    resampled_group = resample(group,
                               replace=True,  # Sample with replacement
                               n_samples=int(target_size),  # Target size for each class
                               random_state=123)  # For reproducibility
    resampled_data = pd.concat([resampled_data, resampled_group])

# Shuffle the resampled data to mix up the order of rows
resampled_data = resampled_data.sample(frac=1, random_state=123).reset_index(drop=True)

# Display the new distribution of 'track_popularity' after resampling
new_target_distribution = resampled_data['track_popularity'].value_counts()

new_target_distribution
```
